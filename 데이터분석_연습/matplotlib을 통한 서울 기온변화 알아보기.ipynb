```python
f = open('seoul.csv') 
data = csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '22':
            result.append(float(row[-1]))

plt.plot(result,'hotpink')
plt.show()
```

```python
f = open('seoul.csv') 
data = csv.reader(f)
next(data)
high=[]
low=[]

for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1983 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '22':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.rc('font',family='Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프')
plt.rcParams['axes.unicode_minus']=False #(-)기호가 깨질 때 추가
plt.legend()
plt.show()
```

```python
f = open('seoul.csv') 
data = csv.reader(f)
next(data)
aug =[]
jan=[]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
        
plt.hist(aug,bins=100,color='r',label='Aug')
plt.hist(jan,bins=100,color='b',label='Jan')
plt.legend()
plt.show()
```

```python
#월별로 구분
f = open('seoul.csv') 
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
```

```python
#8월 일별 기온 데이터를 상자 그림으로 표현하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day=[]
for i in range(31):
    day.append([])
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(day,showfliers=False)

plt.show()
```
